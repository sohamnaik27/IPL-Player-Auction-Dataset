#### Q3] Most expensive top 10 Players

```python
data = ipl_dataset[['Player','Amount','Year']].sort_values("Amount", axis = 0, ascending = False)[:10]
data
```

#### Q4] Most expensive player each year

```python
data = []
for i in sorted(ipl_dataset.Year.unique()):
    df = pd.DataFrame(ipl_dataset.groupby('Year').get_group(i))
    df = df[df['Amount'] == df.Amount.max()]
    data.append([i, np.array(df["Player"]), np.array(df["Amount"])[0], np.array(df["Team"])[0]])
data = pd.DataFrame(data, columns=["Year", "Players", "Amount", "Team"])
data
```

#### Q5] Most expensive player according to their role

```python
data = []
for i in sorted(ipl_dataset.Role.unique()):
    df = pd.DataFrame(ipl_dataset.groupby('Role').get_group(i))
    df = df[df['Amount'] == df.Amount.max()]
    data.append([i, np.array(df["Player"]), np.array(df["Amount"])[0], np.array(df["Team"])[0]])
data = pd.DataFrame(data, columns=["Role", "Player", "Amount", "Team"])
data
```

#### Q6] Most expensive player purchased my each team

```python
data = []
for i in sorted(ipl_dataset.Team.unique()):
    df = pd.DataFrame(ipl_dataset.groupby('Team').get_group(i))
    df = df[df['Amount'] == df.Amount.max()]
    data.append([i, np.array(df["Player"]), np.array(df["Amount"])[0], np.array(df["Team"])[0]])
data = pd.DataFrame(data, columns=["Role", "Player", "Amount", "Team"])
data
```

#### Q7] Higest spending teams each year.

```python
data = []
for i in sorted(ipl_dataset.Year.unique()):
    df = pd.DataFrame(ipl_dataset.groupby('Year').get_group(i))
    df = df[df['Amount'] == df.Amount.max()]
    data.append([i, np.array(df["Amount"])[0], np.array(df["Team"])[0]])
data = pd.DataFrame(data, columns=["Year", "Amount", "Team"])
data
```
